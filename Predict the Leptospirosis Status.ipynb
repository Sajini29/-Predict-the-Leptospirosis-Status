### Preprocessing

### Handling Missing Values

```python
df = df.replace([99,98,'Fiel','occ','99','98','NA'], pd.NA)
```

```python
df.head()
```

```python
# Calculate the missing value percentage for each column
missing_percentages = (df.isnull().sum() / len(df)) * 100

# Print the missing value percentages for each column
print(missing_percentages)
```

```python
# Drop columns with missing value percentage 50% or above
columns_to_drop = missing_percentages[missing_percentages >= 50].index
df.drop(columns=columns_to_drop, inplace=True)
```

```python
df.shape
```

```python
column_list = df.columns.tolist()

print(column_list)
```

```python
for i in column_list:
    column_mode = df[i].mode().values[0]
    df[i] = df[i].fillna(column_mode)
```

```python
# Define the number of rows for training and testing sets

train_rows = 1387
test_rows = 347

#Split the datafrmae into training and testing sets

df11 = df[:train_rows]
df22 = df[train_rows:train_rows+test_rows]
```

### Model Building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier 
from sklearn.svm import SVC
```

```python
smt=SMOTE()
x_sm,y_sm=smt.fit_resample(x,y)
y_sm.value_counts()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_sm, y_sm, test_size=0.2,random_state=2906)
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
importance = rf.feature_importances_
```

```python
feature_importance = pd.DataFrame({'Feature': x_sm.columns, 'Importance': importance})
```

```python
feature_importance = feature_importance.sort_values('Importance', ascending=False)
```

```python
n = 10  # Number of significant variables to select
significant_variables = feature_importance.head(n)['Feature'].tolist()
significant_variables
```

```python
x_train_reduced = x_train[significant_variables]
x_test_reduced = x_test[significant_variables]
```

```python
model_reduced = RandomForestClassifier(random_state=42)
model_reduced.fit(x_train_reduced, y_train)
```

```python
rf_pred = model_reduced.predict(x_test_reduced)
print(classification_report(y_test,rf_pred))
```

```python
from sklearn.neural_network import MLPClassifier
nn = MLPClassifier()
nn.fit(x_train_reduced,y_train)
nn_pred = nn.predict(x_test_reduced)
print(classification_report(y_test,nn_pred))
```

```python
model = LogisticRegression()
model.fit(x_train_reduced,y_train)
y_pred = model.predict(x_test_reduced)
print(classification_report(y_test,y_pred))
```

```python
svc = SVC()
svc.fit(x_train_reduced, y_train)
y_pred_svc = svc.predict(x_train_reduced)
print(classification_report(y_train,y_pred_svc))
```

```python
print(f'logistic reggression: {accuracy_score(y_test,y_pred)}')
print(f'nural network: {accuracy_score(y_test,nn_pred)}')
print(f'random forest: {accuracy_score(y_test,rf_pred)}')
#print(f'xgboost: {accuracy_score(y_test,xgb_pred)}')
```
